**(0.2 балла)** В колонке `trip_duration` записано целевое значение, которое мы хотим предсказывать. Давайте посмотрим на распределение таргета в обучающей выборке. Для этого нарисуйте его гистограмму:

```python
plt.figure(figsize=(11,7))
plt.title("Распределение trip_duration", fontsize=14)
plt.hist(data["trip_duration"], bins=400, color='black', alpha=0.8)
plt.xlabel('Продолжительность поездки (сек)', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.yscale("log")
plt.grid(axis='y', alpha=0.5)
max_val = data["trip_duration"].max()
plt.xticks(np.arange(0, max_val, 100000))
plt.tight_layout()
plt.show()
```

**Вопрос (0.1 балл)**: **(?)**Что можно сказать о целевой переменной по гистограмме её значений?

**Ответ**: Почти все значения сосредоточены от 0 до 100 тысяч секунд, редкие выбросы за миллион секунд, скорее всего, являются ошибочными

В соревновании в качестве метрики качества использовалось RMSLE:
$$\text{RMSLE}(X, y, a) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2}$$

**Вопрос (0.1 балл)**: **(?)**Как вы думаете, почему авторы соревнования выбрали именно RMSLE, а не RMSE?

**Ответ**: мне кажется, что решение выбрать метрику с квадратом разницы логарифмов вместо обычного квадрата разницы значений можно объяснить тем, что авторам соревнования важнее, чтобы модель точно угадывала время коротких поездок. А вот если поездка длительная (условно, больше часа), то там разница в 10-100 секунд не такая критичная. Таким образом, модель сглаживает выбросы в данных и не так сильно зависит от редких длинных поездок. И для этой метрики важнее "относительная" точность, а не абсолютная

На семинаре мы рассматривали несколько моделей линейной регрессии в `sklearn`, но каждая из них оптимизировала среднеквадратичную ошибку (MSE), а не RMSLE. Давайте проделаем следующий трюк: будем предсказывать не целевую переменную, а ее *логарифм*. Обозначим $\hat{y}_i = \log{(y_i + 1)}$ — модифицированный таргет, а $\hat{a}(x_i)$ — предсказание модели, которая обучалась на $\hat{y}_i$, то есть логарифм таргета. Чтобы предсказать исходное значение, мы можем просто взять экспоненту от нашего предсказания: $a(x_i) = \exp(\hat{a}(x_i)) - 1$.

**Вопрос (0.3 балла)**: **(?)**Покажите, что оптимизация RMSLE для модели $a$ эквивалентна оптимизации MSE для модели $\hat{a}$.

**Доказательство**: Для доказательства еще раз перепишем метрику RMSLE и, сделав замены переменных согласно описанию выше ($\hat{y}_i = \log{(y_i + 1)}$ и  $a(x_i) = \exp(\hat{a}(x_i)) - 1$), получим новую метрику.

$\text{RMSLE}(X, y_i, a_i(x_i)) = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\log{(y_i + 1)} - \log{(a(x_i) + 1)}\big)^2} = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\hat{y}_i - \log{(\exp(\hat{a}_i(x_i)))}\big)^2} = \sqrt{\frac{1}{\ell}\sum_{i=1}^{\ell} \big(\hat{y}_i - \hat{a}_i(x_i)\big)^2} = \text{MSE}(X, \hat{y}_i, \hat{a}_i(x_i))$

Хорошо видно, что новая метрика по сути является MSE для модели $\hat{a}$. Эквивалентность моделей доказана


**(0.1 балл)** Итак, мы смогли свести задачу оптимизации RMSLE к задаче оптимизации MSE, которую мы умеем решать! Кроме того, у логарифмирования таргета есть еще одно полезное свойство. Чтобы его увидеть, добавьте к нашей выборке колонку `log_trip_duration` (воспользуйтесь `np.log1p`) и нарисуйте гистограмму модифицированного таргета по обучающей выборке. Удалите колонку со старым таргетом.

```python
data["log_trip_duration"] = np.log1p(data["trip_duration"])

plt.figure(figsize=(11,7))
plt.title("Распределение log_trip_duration", fontsize=14)
plt.hist(data["log_trip_duration"], bins=400, color='black', alpha=0.8)
plt.xlabel('Log от продолжительности поездки', fontsize=12)
plt.ylabel('Частота', fontsize=12)
plt.grid(axis='y', alpha=0.5)
max_val = data["log_trip_duration"].max()
plt.xticks(np.arange(0, max_val, 1))
plt.tight_layout()
plt.show()
```

```python
data_for_fit = data.drop("trip_duration", axis = 1)
data_for_fit.info()
```

```python
best_const_pred = data_for_fit["log_trip_duration"].mean()
mse_best_const_pred = ((best_const_pred - data_for_fit["log_trip_duration"])**2).mean()
mse_best_const_pred
```

```python
def show_circles_on_map(data, latitude_column, longitude_column, color):
    """
    The function draws map with circles on it.
    The center of the map is the mean of coordinates passed in data.

    data: DataFrame that contains columns latitude_column and longitude_column
    latitude_column: string, the name of column for latitude coordinates
    longitude_column: string, the name of column for longitude coordinates
    color: string, the color of circles to be drawn
    """

    location = (data[latitude_column].mean(), data[longitude_column].mean())
    m = folium.Map(location=location)

    for _, row in data.iterrows():
        folium.Circle(
            radius=100,
            location=(row[latitude_column], row[longitude_column]),
            color=color,
            fill_color=color,
            fill=True
        ).add_to(m)

    return m
```

**Задание 9 (1 балл)**. Сейчас мы почти что не используем сами значения координат. На это есть несколько причин: по отдельности рассматривать широту и долготу не имеет особого смысла, стоит рассматривать их вместе. Во-вторых, понятно, что зависимость между нашим таргетом и координатами не линейная. Чтобы как-то использовать координаты, можно прибегнуть к следующему трюку: обрамим область с наибольшим количеством поездок прямоугольником (как на рисунке). Разобьем этот прямоугольник на ячейки. Каждой точке сопоставим номер ее ячейки, а тем точкам, что не попали ни в одну из ячеек, сопоставим значение -1.

Напишите трансформер, который сначала разбивает показанную на рисунке область на ячейки, а затем создает два признака: номер ячейки, в которой началась поездка, и номер ячейки, в которой закончилась поездка. Количество строк и столбцов выберите самостоятельно.

Обратите внимание, что все вычисления должны быть векторизованными, трансформер не должен модифицировать передаваемую ему выборку inplace, а все необходимые статистики (если они вдруг нужны) нужно считать только по обучающей выборке в методе `fit`:

```python
from sklearn.base import BaseEstimator, TransformerMixin

class MapGridTransformer(BaseEstimator, TransformerMixin):
    def __init__(self, n_rows, n_cols, quantile):
        self.n_rows, self.n_cols = n_rows, n_cols
        self.quantile = quantile

    def fit(self, data, y=None):
        all_lats = pd.concat([data['pickup_latitude'], data['dropoff_latitude']])
        all_lons = pd.concat([data['pickup_longitude'], data['dropoff_longitude']])

        alpha = (1 - self.quantile) / 2
        self.lat_border = (all_lats.quantile(alpha), all_lats.quantile(1 - alpha))
        self.lon_border = (all_lons.quantile(alpha), all_lons.quantile(1 - alpha))

        self.lat_cells_ = np.linspace(self.lat_border[0], self.lat_border[1], self.n_rows + 1)
        self.lon_cells_ = np.linspace(self.lon_border[0], self.lon_border[1], self.n_cols + 1)

        return self

    def transform(self, data):
        data_copy = data.copy()
        
        def get_cell_number(lat_series, lon_series):
            lat_idx = np.digitize(lat_series, self.lat_cells_) - 1
            lon_idx = np.digitize(lon_series, self.lon_cells_) - 1
        
            mask_bad_val = ~ ((lat_idx >= 0) & (lat_idx < self.n_rows) &(lon_idx >= 0) & (lon_idx < self.n_cols))
            cell_numbers = lat_idx * self.n_cols + lon_idx
            cell_numbers[mask_bad_val] = -1
        
            return cell_numbers
        
        data_copy['pickup_cell'] = get_cell_number(data_copy['pickup_latitude'], data_copy['pickup_longitude'])
        data_copy['dropoff_cell'] = get_cell_number(data_copy['dropoff_latitude'], data_copy['dropoff_longitude'])
        
        return data_copy

    def get_border(self):
      return self.lat_border, self.lon_border

grid_transformer = MapGridTransformer(n_rows=25, n_cols=25, quantile = 0.9999)
data_for_fit = grid_transformer.fit_transform(data_for_fit)
data_for_fit.info()
```
